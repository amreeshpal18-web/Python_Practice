**Write a program which can remove a particular character from the string**

```python
str1 = "I love Python"
character = input("Enter the character you want to delete: ")
new_str = ""
for i in str1:
  if i != character:
    new_str += i
print("Original string: ", str1)
print("String after removal: ", new_str)
```

**Write a program that check a string is palindrome or not**

```python
str1="naman"
reverse=str1[::-1]
if reverse==str1:
  print("String is Palindrome")
else:
  print("String is not Palindrome")
```

**Write a program to count the number of words WITHOUT split() in a string**

```python
str1="Python is a interpreter based language that means it executes the code line by line"
new_str1=str1.strip()
Count=1
for i in new_str1:
  if i==" ":
    Count+=1
print("Total number of words are :- ",Count)
```
